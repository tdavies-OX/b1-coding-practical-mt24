# Simulate closed-loop
After implementing your control functionality, you can simulate the closed-loop with code that looks something like this...

```python
# Import relevant modules
from pdcontroller import PDController
from dynamic import Submarine, ClosedLoop, Mission

# Define the PD controller gains
kp = 0.15
kd = 0.6

# Create the submarine, controller, closed-loop system, and mission
sub = Submarine()
controller = PDController(kp, kd)
closed_loop = ClosedLoop(sub, controller)
mission = Mission.from_csv("/workspaces/b1-coding-practical-mt24/data/mission.csv")

# Simulate the closed-loop system with random disturbances
trajectory = closed_loop.simulate_with_random_disturbances(mission)
trajectory.plot_completed_mission(mission)
```
